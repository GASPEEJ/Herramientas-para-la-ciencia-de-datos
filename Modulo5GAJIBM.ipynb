```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Cierre de acción')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio de cierre (USD)')
    plt.legend()
    plt.grid()
    plt.show()
```
